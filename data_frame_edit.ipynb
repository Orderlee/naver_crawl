```python
df =df.replace('상세페이지 참조',0)
```

```python
df = df.astype({'기장': float,'소매 기장': float})
```

```python
def all_length(values):
    if 25< values < 48:
        return '숏'
    elif 48< values < 62:
        return '노멀'
    elif values > 62 :
        return '롱'
    else:
        return 0
```

```python
# df['소매 기장'] = df['소매 기장'].apply
```

```python
df['기장'] = df['기장'].apply(all_length)
df
```
