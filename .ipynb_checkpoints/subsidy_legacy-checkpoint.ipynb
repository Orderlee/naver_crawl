```python
def write_spec_csv(region_name, intro_link):
    spec_link = intro_link.replace("intro", "spec")
    file_name = './result/hyundai_result.xlsx'
    start_row = 0
    elems = pd.read_html(spec_link, encoding = "utf-8", )

    # write csv
    os.makedirs("./result", exist_ok = True)
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    if not os.path.exists(file_name):
        writer = pd.ExcelWriter(file_name, engine='openpyxl', mode = "w")
    else:
        writer = pd.ExcelWriter(file_name, engine='openpyxl', mode = "a")
    
    for elem in elems:

        elem.to_excel(writer, sheet_name=car_name, startrow = start_row, encoding = "utf-8-sig")
            
        start_row = start_row + elem.shape[0] + 5
        # dfs = pd.concat([dfs, df])
        
    writer.save()
```
